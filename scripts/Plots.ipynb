# Plotting AUC

```python
import matplotlib.pyplot as plt


%matplotlib inline


datap = [(0, 0), (0, 1), "g", (0, 1), (1, 1), "g"]
datar = [(0, 1), (0, 1), "b"]
dataf = [(0, 1), (0, 0), "r"]
plt.figure(figsize=(5, 5))
plt.title("True Positive Rate as a function of False Positive rate")
plt.xlabel("False Positive Rate")
plt.plot(*datap, label = "Perfect Model")
plt.plot(*datar, label = "Random Model")
plt.plot(*dataf, label = "Worst Model")
plt.savefig("AUC_models.png", dpi=1200, bbox_inches="tight")
plt.show()
```

```python
import numpy as np

N = 100
x = np.linspace(0, 1, 100)
yr = x

datap = [(0, 0), (0, 1), "black", (0, 1), (1, 1), "black"]
datar = [(0, 1), (0, 1), "black"]

fig = plt.figure(figsize=(5, 5))
plt.title("True Positive Rate as a function of False Positive rate")
plt.xlabel("False Positive Rate")
plt.plot(*datap, label="Perfect Model")
plt.plot(*datar, label="Random Model")
plt.fill_between(x, yr, 1, hatch="o", edgecolor="black")
plt.savefig("AUC_models_useful.png", dpi=1200, bbox_inches="tight")
plt.show()
```
